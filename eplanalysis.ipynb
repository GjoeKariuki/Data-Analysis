```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading our data
epl_dframe = pd.read_csv("EPL_20_21.csv")
```

```python
# player minutes played / matches
# player goals scored / matches
epl_dframe['Mins/Match'] = (epl_dframe['Mins'] / epl_dframe['Matches']).astype(int)
epl_dframe['Goals/Match'] = (epl_dframe['Goals'] / epl_dframe['Matches']).astype(float)
```

```python
epl_dframe.head()
```

```python
# total goals scored in the season
total_goals = epl_dframe['Goals'].sum()
```

```python
print(total_goals)
```

```python
# total penalties
total_penaltiescored = epl_dframe['Penalty_Goals'].sum()
```

```python
print(total_penaltiescored)
```

```python
# penalties attempted
total_penaltiesattempted = epl_dframe['Penalty_Attempted'].sum()
```

```python
print(total_penaltiesattempted)
```

```python
missed_penalties = total_penaltiesattempted - total_penaltiescored
```

```python
data = [missed_penalties, total_penaltiescored]
```

```python
# penalties missed / scored 
plt.figure(figsize=(13,6))
color = sns.color_palette("Set2")
plt.pie(data, labels=['Missed Penalties', 'Scored Penalties'],
colors=color, autopct='%.0f%%')
plt.show()
```

```python
# players per club
players_per_club = epl_dframe['Club'].value_counts().sort_values(ascending=False)
```

```python
players_per_club.plot(ylabel='No of players', xlabel='Football Team', kind='bar', color=sns.color_palette('viridis'))
```

```python
# splitting players into age group
under_20 = epl_dframe[epl_dframe['Age'] <= 20]
btw_21_24 = epl_dframe[(epl_dframe['Age'] >= 21) & (epl_dframe['Age'] <= 24)]
btw_25_29 = epl_dframe[(epl_dframe['Age'] >= 25) & (epl_dframe['Age'] <= 29)]
btw_30_34 = epl_dframe[(epl_dframe['Age'] >= 30) & (epl_dframe['Age'] <= 34)]
above_34 = epl_dframe[epl_dframe['Age'] >= 35]
```
