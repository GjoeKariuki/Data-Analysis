```python
all_matches = []
```

```python
epl_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
for year in ([2022,2021,2020]):    
    
    epl_data = requests.get(epl_url)
    epl_soup = BeautifulSoup(epl_data.content)
    standings_table = epl_soup.select('table.stats_table')[0]

    teams_links = [l.get('href') for l in standings_table.find_all('a')]
    teams_links = [l for l in teams_links if '/squads/' in l]
    teams_links = [f'https://fbref.com{l}' for l in teams_links]

    prev_season = epl_soup.select('a.prev')[0].get('href')
    epl_url = f'https://fbref.com/{prev_season}'

    for t in teams_links:
        team_name = ((t.split('/')[-1]).replace('-Stats','')).replace('-',' ')
        team_data = requests.get(t)
        team_matches = pd.read_html(team_data.content, match='Scores & Fixtures')[0]
        team_soup = BeautifulSoup(team_data.content)
        team_link = [ele.get('href') for ele in team_soup.find_all('a')]
        team_link = [ele for ele in team_link if ele and 'all_comps/shooting' in ele]
        search_link = [f'https://fbref.com{e}' for e in team_link]
        shooting_statsdata = requests.get(search_link[0])
        shooting_statdata = pd.read_html(shooting_statsdata.content, match='Shooting')[0]
        shooting_statdata.columns = shooting_statdata.columns.droplevel()

        try:
            team_mergedata = team_matches.merge(shooting_statdata[['Date','Sh','SoT','Dist','FK','PK','PKatt']], on='Date')
        except ValueError:
            continue
        
        team_mergedata = team_mergedata[team_mergedata['Comp'] == 'Premier League']
        team_mergedata['Season'] = year
        team_mergedata['Team'] = team_name
        all_matches.append(team_mergedata)
        time.sleep(3)
```

```python
matches_dframe = pd.concat(all_matches)
```

```python
matches_dframe.columns = [c.lower() for c in matches_dframe.columns]
```

```python
matches_dframe.to_csv('matches_dframe.csv')
```
